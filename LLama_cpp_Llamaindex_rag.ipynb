```python
import os
from llama_index.core.node_parser import SentenceWindowNodeParser
from llama_index.core import VectorStoreIndex, ServiceContext, load_index_from_storage

def get_build_index(documents,llm,embed_model="local:BAAI/bge-small-en-v1.5",sentence_window_size=3,save_dir="./vector_store/index"):

  node_parser = SentenceWindowNodeParser(
      window_size = sentence_window_size,
      window_metadata_key = "window",
      original_text_metadata_key = "original_text"
  )

  sentence_context = ServiceContext.from_defaults(
      chunk_size = 1024,
      llm = llm,
      embed_model= embed_model,
      node_parser = node_parser,
  )

  if not os.path.exists(save_dir):
        # create and load the index
        index = VectorStoreIndex.from_documents(
            [documents], service_context=sentence_context
        )
        index.storage_context.persist(persist_dir=save_dir)
  else:
      # load the existing index
      index = load_index_from_storage(
          StorageContext.from_defaults(persist_dir=save_dir),
          service_context=sentence_context,
      )

  return index
```

```python
from llama_index.core.postprocessor import MetadataReplacementPostProcessor, SentenceTransformerRerank
from llama_index.core.memory import ChatMemoryBuffer

memory = ChatMemoryBuffer.from_defaults(token_limit=1500)

def get_engine(llm, memory, sentence_index, similarity_top_k=3, rerank_top_n=2):
    postproc = MetadataReplacementPostProcessor(target_metadata_key="window")

    rerank = SentenceTransformerRerank(
        top_n=rerank_top_n, model="BAAI/bge-reranker-base"
    )

    engine = sentence_index.as_chat_engine(
        chat_mode = 'context',
        llm = llm,
        memory = memory,
        system_prompt=(
     "Welcome! You're interacting with a chatbot capable of having normal conversations!!! \n"
     "If a question doesn't relate to the given context, just respond with 'There is no mention of the question in the data.' \n"
        # "be proffessional\n"
        # "(((if the querry is meet and greet type like hi!!, hell0!!, how are you!! etc simply ignore and reply casually without searching)))"
        # " about any document as the user chooses. "
        ),
        similarity_top_k = similarity_top_k,
        node_postprocessors = [postproc, rerank],
        streaming=True
    )

    return engine
```

```python
chat_engine = get_engine(llm, memory, sentence_index=vector_index, similarity_top_k=3, rerank_top_n=2)
```

```python
def run_chat_engine(chat_engine):
  while True:
      query = input("User: ")
      if query == "exit":
        break
      else:
        response = chat_engine.stream_chat(query)
        for token in response.response_gen:
            print(token, end="")

        print("\n\n")
run_chat_engine(chat_engine)
```
